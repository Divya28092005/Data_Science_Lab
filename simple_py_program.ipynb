# 1.WAP to compute factorial of a given number

```python
def factorial(n):
    """
    Calculates the factorial of a non-negative integer n.
    """
    if not isinstance(n, int) or n < 0:
        raise ValueError("Input must be a non-negative integer.")
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example usage:
number = 5
try:
    result = factorial(number)
    print(f"The factorial of {number} is {result}")

    number = 0
    result = factorial(number)
    print(f"The factorial of {number} is {result}")

    number = 7
    result = factorial(number)
    print(f"The factorial of {number} is {result}")

    # Example with invalid input
    # number = -3
    # result = factorial(number)
    # print(f"The factorial of {number} is {result}")

    # number = 4.5
    # result = factorial(number)
    # print(f"The factorial of {number} is {result}")
except ValueError as e:
    print(f"Error: {e}")
```

# 2.WAP to display table for a given number

```python
def display_multiplication_table(number, limit=10):
    """
    Displays the multiplication table for a given number up to a specified limit.

    Args:
        number (int): The number for which to display the table.
        limit (int, optional): The upper limit for multiplication. Defaults to 10.
    """
    if not isinstance(number, int) or not isinstance(limit, int) or number < 0 or limit < 1:
        raise ValueError("Number must be a non-negative integer and limit must be a positive integer.")

    print(f"Multiplication Table for {number} up to {limit}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Example usage:
try:
    # Display table for 7 up to 10
    display_multiplication_table(7)
    print("\n") # Add a newline for better readability between tables

    # Display table for 12 up to 5
    display_multiplication_table(12, limit=5)
    print("\n")

    # Display table for 0
    display_multiplication_table(0)
    print("\n")

    # Example with invalid input (uncomment to test error handling)
    # display_multiplication_table(-5)
    # display_multiplication_table(3.5)
    # display_multiplication_table(5, limit=0)
except ValueError as e:
    print(f"Error: {e}")
```

#3.WAP to identify the given number is odd or even


```python
def check_odd_even(number):
    """
    Checks if a given number is odd or even.

    Args:
        number (int): The integer to check.

    Returns:
        str: 'Even' if the number is even, 'Odd' if the number is odd.
    """
    if not isinstance(number, int):
        raise ValueError("Input must be an integer.")

    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Example usage:
try:
    num1 = 4
    print(f"The number {num1} is {check_odd_even(num1)}.")

    num2 = 7
    print(f"The number {num2} is {check_odd_even(num2)}.")

    num3 = 0
    print(f"The number {num3} is {check_odd_even(num3)}.")

    num4 = -5
    print(f"The number {num4} is {check_odd_even(num4)}.")

    # Example with invalid input (uncomment to test error handling)
    # num5 = 3.14
    # print(f"The number {num5} is {check_odd_even(num5)}.")
except ValueError as e:
    print(f"Error: {e}")
```

#4.WAP to identify the entered number is palindrome or not.

```python
def is_palindrome_number(number):
    """
    Checks if a given integer is a palindrome.

    A number is a palindrome if it reads the same forwards and backwards.

    Args:
        number (int): The integer to check.

    Returns:
        bool: True if the number is a palindrome, False otherwise.

    Raises:
        ValueError: If the input is not an integer.
    """
    if not isinstance(number, int):
        raise ValueError("Input must be an integer.")

    # Convert the number to a string to easily compare characters
    num_str = str(number)

    # Compare the string with its reverse
    return num_str == num_str[::-1]

# Example usage:
try:
    num1 = 121
    print(f"Is {num1} a palindrome? {is_palindrome_number(num1)}")

    num2 = 12321
    print(f"Is {num2} a palindrome? {is_palindrome_number(num2)}")

    num3 = 12345
    print(f"Is {num3} a palindrome? {is_palindrome_number(num3)}")

    num4 = -121 # Negative numbers are usually not considered palindromes in this context
    print(f"Is {num4} a palindrome? {is_palindrome_number(num4)}")

    num5 = 0
    print(f"Is {num5} a palindrome? {is_palindrome_number(num5)}")

    # Example with invalid input (uncomment to test error handling)
    # num6 = 12.1
    # print(f"Is {num6} a palindrome? {is_palindrome_number(num6)}")
except ValueError as e:
    print(f"Error: {e}")
```

## Define Expert System Domain and Rules

### Subtask:
Establish the specific domain for the simple decision-making expert system (e.g., a basic troubleshooting guide, a simple recommendation system). Then, define the initial set of rules that govern its decisions using Python data structures like dictionaries or lists of conditions.


### Chosen Domain: Clothing Recommendation System based on Weather Conditions

This expert system will provide clothing recommendations based on two primary weather conditions: temperature (categorized as 'cold', 'moderate', 'warm') and precipitation (categorized as 'none', 'rain', 'snow').

**Reasoning**:
I will define the data structure for the rules using a list of dictionaries and populate it with example clothing recommendations based on weather conditions.



```python
rules = [
    {
        "temperature": "cold",
        "precipitation": "snow",
        "recommendation": "Wear a heavy coat, snow pants, waterproof boots, gloves, and a hat."
    },
    {
        "temperature": "cold",
        "precipitation": "rain",
        "recommendation": "Wear a waterproof, insulated jacket, warm layers, and waterproof boots."
    },
    {
        "temperature": "cold",
        "precipitation": "none",
        "recommendation": "Wear a warm coat, sweater, and warm pants."
    },
    {
        "temperature": "moderate",
        "precipitation": "rain",
        "recommendation": "Wear a light waterproof jacket and comfortable closed-toe shoes."
    },
    {
        "temperature": "moderate",
        "precipitation": "none",
        "recommendation": "Wear a light jacket or sweater and comfortable clothing."
    },
    {
        "temperature": "warm",
        "precipitation": "rain",
        "recommendation": "Wear light, breathable rain gear and sandals or open shoes."
    },
    {
        "temperature": "warm",
        "precipitation": "none",
        "recommendation": "Wear light clothing like a t-shirt and shorts."
    }
]

print("Defined expert system rules:")
for rule in rules:
    print(f"- Temp: {rule['temperature']}, Precip: {rule['precipitation']} -> Rec: {rule['recommendation']}")
```

## Implement Inference Engine

### Subtask:
Write the core Python logic for the inference engine. This component will process user inputs, apply the defined rules from the knowledge base, and derive a conclusion or recommendation.


**Reasoning**:
I need to implement the core logic of the inference engine as a Python function, `get_clothing_recommendation`, that iterates through the defined rules to find a match and returns the corresponding recommendation or a default message.



```python
def get_clothing_recommendation(temperature, precipitation):
    """
    Applies the defined rules to provide a clothing recommendation
    based on temperature and precipitation.

    Args:
        temperature (str): Categorized temperature (e.g., 'cold', 'moderate', 'warm').
        precipitation (str): Categorized precipitation (e.g., 'none', 'rain', 'snow').

    Returns:
        str: A clothing recommendation based on the rules, or a default message.
    """
    for rule in rules:
        if rule["temperature"] == temperature and rule["precipitation"] == precipitation:
            return rule["recommendation"]

    return "No specific recommendation available for these conditions. Please use your best judgment."

# Test cases for the inference engine
print(f"Recommendation for Cold, Snow: {get_clothing_recommendation('cold', 'snow')}")
print(f"Recommendation for Moderate, Rain: {get_clothing_recommendation('moderate', 'rain')}")
print(f"Recommendation for Warm, None: {get_clothing_recommendation('warm', 'none')}")
print(f"Recommendation for Cold, Sunny (no rule): {get_clothing_recommendation('cold', 'sunny')}")
print(f"Recommendation for Hot, None (no rule): {get_clothing_recommendation('hot', 'none')}")
```

## Develop User Interface (Input/Output)

### Subtask:
Create functions to facilitate interaction with the user. This includes collecting necessary information through prompts and clearly presenting the expert system's decision, diagnosis, or recommendation.


**Reasoning**:
I need to implement the user input function, `get_user_input()`, which will prompt the user for temperature and precipitation, and include robust input validation to ensure only predefined categories are accepted.



```python
def get_user_input():
    """
    Prompts the user for temperature and precipitation, with input validation.

    Returns:
        tuple: A tuple containing the validated temperature and precipitation strings.
    """
    valid_temperatures = ['cold', 'moderate', 'warm']
    valid_precipitations = ['none', 'rain', 'snow']

    temperature = ''
    while temperature not in valid_temperatures:
        temperature = input(f"Enter temperature (choices: {', '.join(valid_temperatures)}): ").lower()
        if temperature not in valid_temperatures:
            print("Invalid temperature. Please choose from 'cold', 'moderate', or 'warm'.")

    precipitation = ''
    while precipitation not in valid_precipitations:
        precipitation = input(f"Enter precipitation (choices: {', '.join(valid_precipitations)}): ").lower()
        if precipitation not in valid_precipitations:
            print("Invalid precipitation. Please choose from 'none', 'rain', or 'snow'.")

    return temperature, precipitation

# Reasoning: Now that the input function is defined, I will define the main function to run the expert system.
def run_expert_system():
    """
    Runs the clothing recommendation expert system by collecting user input,
    applying rules, and displaying the recommendation.
    """
    print("\n--- Clothing Recommendation Expert System ---")
    temperature, precipitation = get_user_input()
    recommendation = get_clothing_recommendation(temperature, precipitation)
    print(f"\nBased on the conditions (Temperature: {temperature}, Precipitation: {precipitation}):")
    print(f"Recommendation: {recommendation}")
    print("-------------------------------------------")

# Example of how to run the system:
# run_expert_system()
```
