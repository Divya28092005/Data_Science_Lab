1. Generate a normal distribution and plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# user inputs
mean = float(input("Enter mean value: "))
std_dev = float(input("Enter standard deviation: "))
size = int(input("Enter number of data points: "))

# generate normal distribution
data = np.random.normal(mean, std_dev, size)

# plot histogram
plt.hist(data, bins=30, density=True)

# normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)

plt.plot(x, p)

plt.title("Normal Distribution Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

2. Compute mean, median, variance, and standard deviation for a dataset.

```python
import numpy as np

n = int(input("Enter number of elements: "))

data = []

for i in range(n):
    num = float(input("Enter number: "))
    data.append(num)

mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)

print("Mean =", mean)
print("Median =", median)
print("Variance =", variance)
print("Standard Deviation =", std_dev)
```

3. Simulate a coin toss experiment and compute probabilities.

```python
import numpy as np

tosses = int(input("Enter number of coin tosses: "))

results = np.random.choice(['Head', 'Tail'], tosses)

heads = np.sum(results == 'Head')
tails = np.sum(results == 'Tail')

prob_heads = heads / tosses
prob_tails = tails / tosses

print("Heads =", heads)
print("Tails =", tails)
print("Probability of Heads =", prob_heads)
print("Probability of Tails =", prob_tails)
```
