1. Demonstrate Breadth First Search (BFS)

```python
from collections import deque

graph = {
    'A': ['B','C'],
    'B': ['D','E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

def bfs(graph, start):
    visited = []
    queue = deque([start])

    visited.append(start)

    while queue:
        node = queue.popleft()
        print(node, end=" ")

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)

print("BFS Traversal:")
bfs(graph,'A')
```

2. Demonstrate Depth First Search (DFS)

```python
graph = {
    'A': ['B','C'],
    'B': ['D','E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

def dfs(graph, start):
    visited = []
    stack = [start]

    while stack:
        node = stack.pop()

        if node not in visited:
            print(node, end=" ")
            visited.append(node)

            for neighbor in reversed(graph[node]):
                stack.append(neighbor)

print("DFS Traversal:")
dfs(graph,'A')
```

3. Implementation of A* Algorithm

```python
import heapq

graph = {
'A': {'B':1,'C':3},
'B': {'D':3,'E':1},
'C': {'F':5},
'D': {'G':2},
'E': {'G':1},
'F': {'G':2},
'G': {}
}

heuristic = {
'A':6,'B':4,'C':4,'D':2,'E':2,'F':2,'G':0
}

def astar(start, goal):

    pq = []
    heapq.heappush(pq,(heuristic[start],0,start,[start]))

    while pq:
        f,cost,node,path = heapq.heappop(pq)

        if node == goal:
            print("Shortest Path:",path)
            print("Total Cost:",cost)
            return

        for neighbor,weight in graph[node].items():
            new_cost = cost + weight
            heapq.heappush(pq,(new_cost + heuristic[neighbor],new_cost,neighbor,path+[neighbor]))

astar('A','G')
```

4. Basic Minimax Algorithm

```python
import math

game_tree = {
    'Root': {
        'A': {'A1':3,'A2':5},
        'B': {'B1':2,'B2':9}
    }
}

def minimax(node, maximizing):

    if isinstance(node,int):
        return node

    if maximizing:
        best = -math.inf
        for child in node.values():
            val = minimax(child, False)
            best = max(best,val)
        return best
    else:
        best = math.inf
        for child in node.values():
            val = minimax(child, True)
            best = min(best,val)
        return best

print("Optimal Value:", minimax(game_tree['Root'], True))
```
