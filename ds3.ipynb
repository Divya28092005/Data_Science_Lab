a) Create a 3×3 Matrix and Compute its Determinant (User Input)

```python
import numpy as np

print("Enter elements of 3x3 matrix:")

matrix = []
for i in range(3):
    row = list(map(int, input(f"Enter row {i+1} elements separated by space: ").split()))
    matrix.append(row)

matrix = np.array(matrix)

print("Original 3x3 matrix:")
print(matrix)

determinant = np.linalg.det(matrix)

print("\nDeterminant of the matrix:", determinant)
```

b) Find the Inverse of the Matrix (User Input)

```python
import numpy as np

print("Enter elements of 3x3 matrix:")

matrix = []
for i in range(3):
    row = list(map(int, input(f"Enter row {i+1} elements separated by space: ").split()))
    matrix.append(row)

matrix = np.array(matrix)

print("Matrix:")
print(matrix)

det = np.linalg.det(matrix)

print("\nDeterminant:", det)

if det != 0:
    inverse_matrix = np.linalg.inv(matrix)
    print("\nInverse of the matrix:")
    print(inverse_matrix)
else:
    print("\nMatrix is singular, inverse does not exist.")
```

c) Solve Linear Equations using NumPy (User Input)

```python
import numpy as np

print("Enter coefficients for equation 1 (a b):")
a1, b1 = map(int, input().split())

print("Enter coefficients for equation 2 (a b):")
a2, b2 = map(int, input().split())

print("Enter constants for equations:")
c1, c2 = map(int, input().split())

A = np.array([[a1, b1],
              [a2, b2]])

B = np.array([c1, c2])

print("Coefficient Matrix A:")
print(A)

print("\nConstants Vector B:")
print(B)

solution = np.linalg.solve(A, B)

print("\nSolution for [x, y]:")
print(solution)
```
