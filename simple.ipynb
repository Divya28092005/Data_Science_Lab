### Factorial Calculation

The following code defines a recursive function `factorial(n)` that calculates the factorial of a non-negative integer `n`. The factorial of a number `n` is the product of all positive integers less than or equal to `n`.

```python
def factorial(n):
    """
    Computes the factorial of a non-negative integer.

    Args:
        n: An integer.

    Returns:
        The factorial of n.

    Raises:
        ValueError: If n is a negative integer.
    """
    if not isinstance(n, int):
        raise TypeError("Input must be an integer.")
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    elif n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}")

number_two = 0
result_two = factorial(number_two)
print(f"The factorial of {number_two} is {result_two}")

try:
    factorial(-1)
except ValueError as e:
    print(f"Error: {e}")

try:
    factorial(3.5)
except TypeError as e:
    print(f"Error: {e}")
```

# Display_Multiplication_Table



```python
def display_multiplication_table(number):
    """
    Displays the multiplication table for a given number.

    Args:
        number: An integer for which the multiplication table will be displayed.
    """
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11): # Generates table from 1 to 10
        print(f"{number} x {i} = {number * i}")

# Example usage:
number_to_table = 7
display_multiplication_table(number_to_table)

print("\n") # Add a newline for better readability between tables

number_to_table_two = 12
display_multiplication_table(number_to_table_two)
```

```python
def check_odd_even(number):
    """
    Checks if a given number is odd or even.

    Args:
        number: An integer.

    Returns:
        A string indicating whether the number is 'Even' or 'Odd'.

    Raises:
        TypeError: If the input is not an integer.
    """
    if not isinstance(number, int):
        raise TypeError("Input must be an integer.")
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Example usage:
number_one = 10
print(f"The number {number_one} is {check_odd_even(number_one)}.")

number_two = 7
print(f"The number {number_two} is {check_odd_even(number_two)}.")

number_three = 0
print(f"The number {number_three} is {check_odd_even(number_three)}.")

try:
    check_odd_even(4.5)
except TypeError as e:
    print(f"Error: {e}")
```

# Plindrome program

```python
def is_palindrome(number):
    """
    Checks if a given number is a palindrome.

    Args:
        number: An integer.

    Returns:
        True if the number is a palindrome, False otherwise.

    Raises:
        TypeError: If the input is not an integer.
    """
    if not isinstance(number, int):
        raise TypeError("Input must be an integer.")

    # Convert the number to a string to easily reverse it
    num_str = str(number)

    # Compare the original string with its reversed version
    return num_str == num_str[::-1]

# Example usage:
print(f"Is 121 a palindrome? {is_palindrome(121)}")
print(f"Is 12321 a palindrome? {is_palindrome(12321)}")
print(f"Is 12345 a palindrome? {is_palindrome(12345)}")
print(f"Is 0 a palindrome? {is_palindrome(0)}")
print(f"Is 11 a palindrome? {is_palindrome(11)}")
print(f"Is -121 a palindrome? {is_palindrome(-121)}") # Note: typically palindromes are non-negative

try:
    is_palindrome(12.34)
except TypeError as e:
    print(f"Error: {e}")
```

## Define Expert System Concept

### What is an Expert System?
An **Expert System** is a computer program that simulates the judgment and behavior of a human expert in a specific domain. Its primary goal is to solve complex problems or provide advice by reasoning through a body of knowledge, represented primarily as if-then rules rather than through conventional procedural code.

### Key Components of an Expert System:
1.  **Knowledge Base:** This is the core component that stores domain-specific knowledge acquired from human experts. It typically consists of facts and, crucially for our purpose, a set of **rules** (often in an "if-then" format) that define relationships and conditions within the domain.
2.  **Inference Engine:** This component acts as the 'brain' of the expert system. It processes the knowledge base to draw conclusions and make recommendations. For **rule-based decision-making**, the inference engine uses logical rules to deduce new facts or actions from existing data. It determines which rules to apply and in what order.
3.  **Working Memory (or Fact Base):** This temporary storage holds the specific facts about the current problem being solved by the expert system. It's where the input data is placed and where intermediate conclusions are stored.
4.  **User Interface:** This allows the user to interact with the system, input problem-specific data, and receive advice or solutions.
5.  **Explanation Facility (Optional but common):** This module explains the reasoning process to the user, detailing how the system arrived at a particular conclusion, which is crucial for trust and debugging.
6.  **Knowledge Acquisition Facility (Optional):** This helps in acquiring knowledge from human experts and integrating it into the knowledge base.

### Rule-Based Decision-Making:
Our focus will be on **rule-based decision-making**. In this paradigm, the expert system's intelligence comes from a collection of explicit rules defined by human experts. The inference engine then applies these rules to the available facts to reach a decision. For example, a simple rule might be: "IF `temperature` IS `high` AND `symptoms` INCLUDE `fever` THEN `diagnosis` IS `flu`." This approach is transparent, making the system's logic easy to understand and audit, which is a significant advantage in many applications.

## Define Decision-Making Rules

Establish a set of IF-THEN rules for clothing recommendations based on weather conditions (temperature and precipitation).


### Clothing Recommendation Rules

Based on temperature and precipitation, the following IF-THEN rules are established for clothing recommendations:

*   **IF Temperature is Cold (< 10°C) AND it is Raining:**
    *   **THEN:** Recommend a heavy waterproof coat, warm layers, waterproof pants, hat, and gloves.

*   **IF Temperature is Cold (< 10°C) AND it is NOT Raining:**
    *   **THEN:** Recommend a heavy coat, warm layers (sweater, long-sleeved shirt), and a hat.

*   **IF Temperature is Moderate (10°C - 25°C) AND it is Raining:**
    *   **THEN:** Recommend a light waterproof jacket, a light sweater or long-sleeved shirt, and an umbrella.

*   **IF Temperature is Moderate (10°C - 25°C) AND it is NOT Raining:**
    *   **THEN:** Recommend a light jacket or cardigan, a t-shirt, and long pants.

*   **IF Temperature is Hot (> 25°C) AND it is Raining:**
    *   **THEN:** Recommend light, breathable clothing (e.g., shorts, t-shirt), a light waterproof jacket, and an umbrella.

*   **IF Temperature is Hot (> 25°C) AND it is NOT Raining:**
    *   **THEN:** Recommend light, breathable clothing (e.g., shorts, t-shirt, sundress), and sunscreen.

## Implement Rule Engine

### Subtask:
Develop Python code that stores the defined rules and an inference engine capable of processing user inputs to arrive at a decision based on these rules.


**Reasoning**:
The subtask requires defining rules, creating a Python class for a clothing expert system, implementing a recommendation method, and testing it with examples. This code block will cover all these requirements by defining the rules as a list of dictionaries, creating the `ClothingExpertSystem` class with the `__init__` and `get_recommendation` methods, and then demonstrating its usage with example inputs.



```python
rules = [
    # Rules for temperatures below 10°C
    {
        'min_temp': -float('inf'),
        'max_temp': 9.9,
        'is_raining': True,
        'recommendation': 'heavy waterproof coat, warm layers, waterproof pants, hat, and gloves'
    },
    {
        'min_temp': -float('inf'),
        'max_temp': 9.9,
        'is_raining': False,
        'recommendation': 'heavy winter coat, warm layers, hat, and gloves'
    },
    # Rules for temperatures between 10°C and 20°C
    {
        'min_temp': 10,
        'max_temp': 19.9,
        'is_raining': True,
        'recommendation': 'waterproof jacket, light layers, and a hat'
    },
    {
        'min_temp': 10,
        'max_temp': 19.9,
        'is_raining': False,
        'recommendation': 'light jacket or sweater, and long pants'
    },
    # Rules for temperatures between 20°C and 30°C
    {
        'min_temp': 20,
        'max_temp': 29.9,
        'is_raining': True,
        'recommendation': 'light waterproof jacket or umbrella, and light clothing'
    },
    {
        'min_temp': 20,
        'max_temp': 29.9,
        'is_raining': False,
        'recommendation': 't-shirt, shorts or light pants, and sunglasses'
    },
    # Rules for temperatures 30°C and above
    {
        'min_temp': 30,
        'max_temp': float('inf'),
        'is_raining': True,
        'recommendation': 'light, breathable waterproof clothing, and stay hydrated'
    },
    {
        'min_temp': 30,
        'max_temp': float('inf'),
        'is_raining': False,
        'recommendation': 'very light clothing, wide-brimmed hat, sunscreen, and stay hydrated'
    }
]

class ClothingExpertSystem:
    def __init__(self, rules):
        self.rules = rules

    def get_recommendation(self, temperature, is_raining):
        if not isinstance(temperature, (int, float)) or not isinstance(is_raining, bool):
            raise TypeError("Temperature must be a number and is_raining must be a boolean.")

        for rule in self.rules:
            if rule['min_temp'] <= temperature < rule['max_temp'] and \
               rule['is_raining'] == is_raining:
                return rule['recommendation']
        return 'No specific recommendation found for these conditions.'

# Instantiate the expert system
expert_system = ClothingExpertSystem(rules)

# Test cases
print(f"Recommendation for 5°C and raining: {expert_system.get_recommendation(5, True)}")
print(f"Recommendation for 5°C and not raining: {expert_system.get_recommendation(5, False)}")
print(f"Recommendation for 15°C and raining: {expert_system.get_recommendation(15, True)}")
print(f"Recommendation for 15°C and not raining: {expert_system.get_recommendation(15, False)}")
print(f"Recommendation for 25°C and raining: {expert_system.get_recommendation(25, True)}")
print(f"Recommendation for 25°C and not raining: {expert_system.get_recommendation(25, False)}")
print(f"Recommendation for 32°C and raining: {expert_system.get_recommendation(32, True)}")
print(f"Recommendation for 32°C and not raining: {expert_system.get_recommendation(32, False)}")

# Edge cases / out of bounds (though our rules are comprehensive)
print(f"Recommendation for -2°C and raining: {expert_system.get_recommendation(-2, True)}")
print(f"Recommendation for 40°C and not raining: {expert_system.get_recommendation(40, False)}")

# Test with invalid input type
try:
    expert_system.get_recommendation('20', False)
except TypeError as e:
    print(f"Error: {e}")
```

## Create User Interface

### Subtask:
Implement a simple command-line interface to allow users to input relevant information (e.g., temperature, rain) and display the expert system's recommendations or decisions.


**Reasoning**:
The subtask requires creating a command-line interface. This code block will implement the CLI with input prompts, error handling, calls to the expert system, and an exit option, as specified in the instructions.



```python
print("\n--- Clothing Recommendation Expert System ---")
print("Enter weather conditions to get a clothing recommendation.")
print("Type 'exit' at any prompt to quit.\n")

while True:
    temp_input = input("Enter the current temperature in Celsius (e.g., 20): ").strip().lower()
    if temp_input == 'exit':
        break

    try:
        temperature = float(temp_input)
    except ValueError:
        print("Invalid temperature input. Please enter a number.")
        continue

    rain_input = input("Is it raining? (yes/no): ").strip().lower()
    if rain_input == 'exit':
        break

    is_raining = None
    if rain_input == 'yes':
        is_raining = True
    elif rain_input == 'no':
        is_raining = False
    else:
        print("Invalid input for rain. Please enter 'yes' or 'no'.")
        continue

    try:
        recommendation = expert_system.get_recommendation(temperature, is_raining)
        print(f"\nRecommendation: Wear {recommendation}.\n")
    except TypeError as e:
        print(f"An internal error occurred: {e}. Please check inputs.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

print("Thank you for using the Clothing Recommendation Expert System!")
```
