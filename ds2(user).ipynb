1. Breadth First Search (BFS)

```python
from collections import deque

graph = {}

n = int(input("Enter number of nodes: "))

for i in range(n):
    node = input("Enter node: ")
    neighbors = input("Enter neighbors separated by space: ").split()
    graph[node] = neighbors

start = input("Enter starting node: ")

def bfs(graph, start):
    visited = []
    queue = deque([start])
    visited.append(start)

    while queue:
        node = queue.popleft()
        print(node, end=" ")

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)

print("BFS Traversal:")
bfs(graph, start)
```

2. Depth First Search (DFS)

```python
graph = {}

n = int(input("Enter number of nodes: "))

for i in range(n):
    node = input("Enter node: ")
    neighbors = input("Enter neighbors separated by space: ").split()
    graph[node] = neighbors

start = input("Enter starting node: ")

def dfs(graph, start):
    visited = []
    stack = [start]

    while stack:
        node = stack.pop()

        if node not in visited:
            print(node, end=" ")
            visited.append(node)

            for neighbor in reversed(graph[node]):
                stack.append(neighbor)

print("DFS Traversal:")
dfs(graph, start)
```

3. A* Algorithm

```python
import heapq

graph = {}
heuristic = {}

n = int(input("Enter number of nodes: "))

for i in range(n):
    node = input("Enter node: ")
    neighbors = {}

    m = int(input("Enter number of neighbors: "))

    for j in range(m):
        neighbor = input("Neighbor name: ")
        cost = int(input("Cost: "))
        neighbors[neighbor] = cost

    graph[node] = neighbors

for node in graph:
    h = int(input("Enter heuristic value for " + node + ": "))
    heuristic[node] = h

start = input("Enter start node: ")
goal = input("Enter goal node: ")

def astar(start, goal):

    pq = []
    heapq.heappush(pq, (heuristic[start], 0, start, [start]))

    while pq:
        f, cost, node, path = heapq.heappop(pq)

        if node == goal:
            print("Shortest Path:", path)
            print("Total Cost:", cost)
            return

        for neighbor, weight in graph[node].items():
            new_cost = cost + weight
            heapq.heappush(pq, (new_cost + heuristic[neighbor], new_cost, neighbor, path + [neighbor]))

astar(start, goal)
```

4. Basic Minimax Algorithm

```python
import math

a1 = int(input("Enter value for A1: "))
a2 = int(input("Enter value for A2: "))
b1 = int(input("Enter value for B1: "))
b2 = int(input("Enter value for B2: "))

game_tree = {
    'Root': {
        'A': {'A1': a1, 'A2': a2},
        'B': {'B1': b1, 'B2': b2}
    }
}

def minimax(node, maximizing):

    if isinstance(node, int):
        return node

    if maximizing:
        best = -math.inf
        for child in node.values():
            val = minimax(child, False)
            best = max(best, val)
        return best
    else:
        best = math.inf
        for child in node.values():
            val = minimax(child, True)
            best = min(best, val)
        return best

print("Optimal Value:", minimax(game_tree['Root'], True))
```
