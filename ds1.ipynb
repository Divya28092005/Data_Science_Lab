1. WAP to Compute Factorial of a Given Number

```python
# Program to find factorial of a number

num = int(input("Enter a number: "))

factorial = 1

if num < 0:
    print("Factorial is not defined for negative numbers")
else:
    for i in range(1, num + 1):
        factorial = factorial * i

    print("Factorial of", num, "is", factorial)
```

2. WAP to Display Table for a Given Number

```python
# Program to print multiplication table

num = int(input("Enter a number: "))

print("Multiplication Table of", num)

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

3. WAP to Identify the Given Number is Odd or Even

```python
# Program to check odd or even

num = int(input("Enter a number: "))

if num % 2 == 0:
    print(num, "is Even")
else:
    print(num, "is Odd")
```

4. WAP to Check Whether a Number is Palindrome or Not

```python
# Program to check palindrome number

num = int(input("Enter a number: "))

temp = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if temp == reverse:
    print(temp, "is a Palindrome")
else:
    print(temp, "is not a Palindrome")
```

5. Expert System for Simple Decision Making

Program 1: Rule Engine (Clothing Expert System)

```python
# Rule based expert system

rules = [
    {'min_temp': -100, 'max_temp': 9, 'rain': True,
     'rec': 'Wear heavy waterproof coat, warm clothes, hat and gloves'},

    {'min_temp': -100, 'max_temp': 9, 'rain': False,
     'rec': 'Wear heavy coat and warm clothes'},

    {'min_temp': 10, 'max_temp': 25, 'rain': True,
     'rec': 'Wear light waterproof jacket and carry umbrella'},

    {'min_temp': 10, 'max_temp': 25, 'rain': False,
     'rec': 'Wear light jacket and long pants'},

    {'min_temp': 26, 'max_temp': 100, 'rain': True,
     'rec': 'Wear light clothes and carry umbrella'},

    {'min_temp': 26, 'max_temp': 100, 'rain': False,
     'rec': 'Wear light clothes like t-shirt and shorts'}
]


def get_recommendation(temp, rain):
    for rule in rules:
        if rule['min_temp'] <= temp <= rule['max_temp'] and rule['rain'] == rain:
            return rule['rec']


# Testing rules
print(get_recommendation(5, True))
print(get_recommendation(15, False))
print(get_recommendation(30, False))
```

Program 2: User Interface for Expert System


```python
# Clothing recommendation expert system

while True:

    temp = input("Enter temperature in Celsius (or type exit): ")

    if temp == "exit":
        break

    temp = float(temp)

    rain = input("Is it raining? (yes/no): ")

    if rain == "yes":
        rain = True
    else:
        rain = False

    if temp < 10 and rain == True:
        print("Wear heavy waterproof coat, warm clothes, hat and gloves")

    elif temp < 10 and rain == False:
        print("Wear heavy coat and warm clothes")

    elif temp >= 10 and temp <= 25 and rain == True:
        print("Wear light waterproof jacket and carry umbrella")

    elif temp >= 10 and temp <= 25 and rain == False:
        print("Wear light jacket and long pants")

    elif temp > 25 and rain == True:
        print("Wear light clothes and carry umbrella")

    elif temp > 25 and rain == False:
        print("Wear light clothes like t-shirt and shorts")
```
